```python
df = pd.read_csv('../datasets/iris.csv')
df
```

```python
x,y = df.drop('Species', axis = 1), df['Species']
x_train,x_test, y_train, y_test = train_test_split(x,y,random_state=25, test_size=0.3, stratify=y)
```

```python
scaler = StandardScaler().set_output(transform = 'pandas')
x_trn_scaled = scaler.fit_transform(x_train)
x_tst_scaled = scaler.transform(x_test)
```

```python
prcomp = PCA() .set_output(transform = 'pandas')
x_trn_pca = prcomp.fit_transform(x_trn_scaled)
x_tst_pca = prcomp.transform(x_tst_scaled)
```

```python
x_trn_pca
```

```python
lr = LogisticRegression()
lr.fit(x_trn_pca.iloc[:,:2],y_train)
```

```python
y_pred = lr.predict(x_tst_pca.iloc[:,:2])
y_pred_proba = lr.predict_proba(x_tst_pca.iloc[:,:2])
```

```python
print(f"Accuracy score is {accuracy_score(y_test, y_pred)}")
print(f"Log Loss is {log_loss(y_test, y_pred_proba)}")
```

# On HR dataset

```python
hr = pd.read_csv('../datasets/HR_comma_sep.csv')
hr
```

```python
x,y = hr.drop('left',axis=1),hr['left']
```

```python
x
```

```python
ohe = OneHotEncoder(drop = 'first', sparse_output=False).set_output(transform = 'pandas')
col_trnf = ColumnTransformer([('OHE', ohe, make_column_selector(dtype_include = object ))],
                             remainder = 'passthrough',
                             verbose_feature_names_out=False)
col_trnf = col_trnf.set_output(transform = 'pandas')

x = col_trnf.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=25, stratify=y, test_size=0.3)
```

```python
scaler = StandardScaler()

x_scaled = scaler.fit_transform(x)
```

```python
pca = PCA().set_output(transform = 'pandas')
x_pca = pca.fit_transform(x_scaled)
x_pca
```

```python
np.cumsum(pca.explained_variance_ratio_ * 100)
```

```python
lr = LogisticRegression()
scores = []
for n in np.arange(8,17):
    # lr.fit_transform(x_pca.iloc[:,:n], y_)
    # print(n)
    prcomp = PCA(n_components=n).set_output(transform = 'pandas')
    pipe = Pipeline([('Scaling', scaler), ('PCA', prcomp), ('Logistic Regression', lr)])
    pipe.fit(x_train, y_train)
    y_pred_proba = pipe.predict_proba(x_test)
    y_pred = pipe.predict(x_test)

    scores.append([n, log_loss(y_test, y_pred_proba), accuracy_score(y_test, y_pred)])

scores_Df = pd.DataFrame(data = scores, columns = ['No. of components', 'Log Loss', 'Accuracy'])
scores_Df.sort_values('Log Loss', ascending=True)
```

using percentage n_components as pecentage

```python
lr = LogisticRegression()
scores = []
for n in np.linspace(0.5, 0.95, 15):
    prcomp = PCA(n_components=n).set_output(transform = 'pandas')
    pipe = Pipeline([('Scaling', scaler), ('PCA', prcomp), ('Logistic Regression', lr)])
    pipe.fit(x_train, y_train)
    y_pred_proba = pipe.predict_proba(x_test)
    y_pred = pipe.predict(x_test)

    scores.append([n*100, log_loss(y_test, y_pred_proba), accuracy_score(y_test, y_pred)])

scores_Df = pd.DataFrame(data = scores, columns = ['precentage variance', 'Log Loss', 'Accuracy'])
scores_Df.sort_values('Log Loss', ascending=True)
```
