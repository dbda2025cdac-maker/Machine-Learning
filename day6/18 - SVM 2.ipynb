```python
os.chdir('C:/Users/dbda.STUDENTSDC/Desktop/M.L/Cases')
vehicle = pd.read_csv('Vehicle_Silhouettes/Vehicle.csv')

vehicle
```

```python
x, y = vehicle.drop('Class', axis = 1), vehicle.Class

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=25,stratify=y)

# from sklearn.compose import ColumnTransformer, make_column_selector

# col_trnf = ColumnTransformer([('OHE', ohe, make_column_selector(dtype_include=object))],
#                               remainder='passthrough', verbose_feature_names_out=False)
```

```python
# For Different Cs and different decision function shapes in Linear SVM
ss = StandardScaler()
dfs=['ovo', 'ovr']
score = []

for c in tqdm(np.linspace(0.01, 5, 15)):
    for s in dfs:
        svm = SVC(kernel='linear', C=c, decision_function_shape=s)
        pipe = Pipeline([('SS', ss), ('SVM', svm)])
        pipe.fit(x_train,y_train)
        y_pred = pipe.predict(x_test)
        # print('For C = ', c)
        # print('For degree = ', d)
        # print('Accuracy Score: ', accuracy_score(y_test, y_pred))
        # print('\n')
        score.append([c,s,accuracy_score(y_test, y_pred)])

df_score = pd.DataFrame(score, columns = ['C', 'DFS', 'Accuracy Score'])
df_score.sort_values('Accuracy Score', ascending=False)
```

```python
# For Different Cs and different degrees in Poly SVM
ss = StandardScaler()
dfs=['ovo', 'ovr']
degrees = [2,3,4]
score = []

for c in tqdm(np.linspace(0.01, 5, 15)):
    for d in degrees:
        for s in dfs:
            svm = SVC(kernel='poly', C=c, degree=d, decision_function_shape=s)
            pipe = Pipeline([('SS', ss), ('SVM', svm)])
            pipe.fit(x_train,y_train)
            y_pred = pipe.predict(x_test)
            # print('For C = ', c)
            # print('For degree = ', d)
            # print('Accuracy Score: ', accuracy_score(y_test, y_pred))
            # print('\n')
            score.append([c,d,s,accuracy_score(y_test, y_pred)])

df_score = pd.DataFrame(score, columns = ['C', 'D', 'DFS', 'Accuracy Score'])
df_score.sort_values('Accuracy Score', ascending=False)
```

```python
# For Different Cs, gammas and decision function shapes in RBF SVM
ss = StandardScaler()
dfs=['ovo', 'ovr']
score = []

for c in tqdm(np.linspace(0.01, 5, 15)):
    for s in dfs:
        for g in np.linspace(0.01, 5, 10):
            svm = SVC(kernel='rbf', C=c, decision_function_shape=s, gamma=g)
            pipe = Pipeline([('SS', ss), ('SVM', svm)])
            pipe.fit(x_train,y_train)
            y_pred = pipe.predict(x_test)
            # print('For C = ', c)
            # print('For degree = ', d)
            # print('Accuracy Score: ', accuracy_score(y_test, y_pred))
            # print('\n')
            score.append([c,g,s,accuracy_score(y_test, y_pred)])

df_score = pd.DataFrame(score, columns = ['C', 'Gamma','DFS', 'Accuracy Score'])
df_score.sort_values('Accuracy Score', ascending=False)
```
